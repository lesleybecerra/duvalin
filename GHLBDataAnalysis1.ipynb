```python
#3 Distrubution of  variables

df_star98[['LOWINC','NABOVE']].hist(bins=20)
df_star98[['LOWINC','NABOVE']].plot.hist(bins=20, alpha =0.5)
```

```python
yax = df_star98.groupby(['LOWINC'])['NABOVE'].mean()
plt.barh(yax.index, yax)
```

```python
#Comparison of summary statistics

subsetd = df_star98[['LOWINC','NABOVE']]
pd.plotting.scatter_matrix(subsetd, alpha=0.9, marker="o", figsize=(15,15),diagonal = 'kde', grid=True)
```
